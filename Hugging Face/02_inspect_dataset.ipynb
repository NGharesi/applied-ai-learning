## Goal
Understand the structure, features, and content of a dataset before applying machine learning models.

## Key Concept
Before training any model, it is critical to explore the dataset to understand:
- data types
- feature structure
- label distribution
- potential issues (missing values, imbalance)

```python
from datasets import load_dataset
from collections import Counter


dataset = load_dataset("imdb")

# Inspect one sample
print(dataset["train"][0])

# Inspect features
dataset["train"].features

# Check dataset size
print(len(dataset["train"]))
print(len(dataset["test"]))

# Check label distribution
labels = dataset["train"][:1000]["label"]
print(Counter(labels))
```

## What I learned
- Datasets have structured features and predefined splits
- Understanding data format is essential before preprocessing
- Label distribution helps identify potential bias or imbalance
