## Goal
Prepare raw dataset by cleaning and transforming the data.

## Key Concept
Typical steps include:
- filtering unwanted samples
- transforming text
- tokenization (later step)
- feature preparation

```python
from datasets import load_dataset
dataset = load_dataset("imdb")

## Filter short texts
dataset = dataset.filter(lambda x: len(x["text"]) > 50)

## Lowercase text
def lowercase(example):
    example["text"] = example["text"].lower()
    return example

dataset = dataset.map(lowercase)
```

## What I learned
- preprocessing transforms raw data into usable inputs
- filtering improves data quality
- mapping functions allow scalable transformations
